```python
import gensim
from gensim.parsing.preprocessing import remove_stopwords
from gensim.parsing.preprocessing import strip_non_alphanum
from nltk import word_tokenize
import pandas as pd
```

```python
tokenized_sentences = []
```

```python
df = pd.read_csv("../LSTACode/lsta_metadata.tsv", sep='\t')
```

```python
df.columns
```

```python
tokenized_sentences = []
for t in df["Ocr text"]:
    if isinstance(t, str):
        t = strip_non_alphanum(t)
        t = t.lower()
        t = remove_stopwords(t)
        tokenized_sentences.append(word_tokenize(t))
    
```

```python
model = gensim.models.Word2Vec(tokenized_sentences, size=100, window=5, min_count=1, 
                               sg=1, alpha=0.025, iter=5, batch_words=10000, workers=1,
                              max_vocab_size=10000)
```

```python
print(len(model.wv.index2word))
print(model.wv.index2word)
```

```python
model.wv.most_similar('aids')
```

```python
model.wv.most_similar('risk')
```
